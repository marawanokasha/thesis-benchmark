## Gzip preprocessed_data files

```python
import shutil
import gzip
import os
from thesis.utils.file import *
```

```python
root_location = "/mnt/virtual-machines/data/"

preprocessed_location = root_location + "preprocessed_data/extended_pv_abs_desc_claims_full_chunks/"
preprocessed_gzip_location = root_location + "preprocessed_data_gzip/extended_pv_abs_desc_claims_full_chunks/"

ensure_disk_location_exists(preprocessed_gzip_location)
```

```python
GZIP_EXTENSION = ".gz"
```

```python
for f in natural_sort(os.listdir(preprocessed_location)):
    print f
```

```python
for f in natural_sort(os.listdir(preprocessed_location)):
    print f
    with open(os.path.join(preprocessed_location, f), 'rb') as f_in, gzip.open(os.path.join(preprocessed_gzip_location, f), 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

## Move gzipped files to the preprocessed_data folder with a .gz extension

```python
for f in natural_sort(os.listdir(preprocessed_gzip_location)):
    print f
    shutil.move(os.path.join(preprocessed_gzip_location, f), os.path.join(preprocessed_location, f + GZIP_EXTENSION))
```
